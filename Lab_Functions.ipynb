## 6. Write a function that returns the factorial of a number.

```python
def factorial(n):
    if not n:
        return None

    if n<=0:
        return "Introuce a positive number"

    elif n>0:

        total_multiplied = 1
    
        while n>1 :
      
         total_multiplied =  total_multiplied * n
         n=n-1

        return total_multiplied

    
#your code here
```

## 11. Write a function that receives a string of comma separated words and returns a string of comma separated words sorted alphabetically.

`NOTE: You may use sorted but not split and definitely no join! ðŸ¤ª`

```python
def sort_alpha(string):
#your code here


    words = []
    word = ""
    
    
    for char in string:
        if char == ',':
            if word:
                words.append(word.strip())
                word = ""
        else:
            word += char
    if word:
        words.append(word.strip())
    
    # Sort the list of words alphabetically
    for i in range(len(words)):
        for j in range(i + 1, len(words)):
            if words[i] > words[j]:
                words[i], words[j] = words[j], words[i]

    # Manually join the sorted words into a comma-separated string
    sorted_words = ""
    for i in range(len(words)):
        sorted_words += words[i]
        if i < len(words) - 1:
            sorted_words += ", "
    
    return sorted_words
```

```python
# This will test your function 
test_alpha(sort_alpha)
```

## 12. Write a function to check if a given password is strong (at least 8 characters, at least one lower case, at least one upper case, at least one number and at least one special character). It should output True if strong and False if not.

```python
def check_pass(password):
#your code here

    
    has_lower = False
    has_upper = False
    has_digit = False
    has_special = False
    special_characters = "@$!%*?&#_-,;'{}()[]â‚¬$"

    # Check the length of the password
    length = 0
    for char in password:
        length += 1

    if length < 8:
        return False

    # Check for at least one lowercase letter, one uppercase letter, one digit, and one special character
    for char in password:
        if char.islower():
            has_lower = True
        elif char.isupper():
            has_upper = True
        elif char.isdigit():
            has_digit = True
        elif char in special_characters:
            has_special = True

    # Return True if all conditions are met, else False
    if has_lower and has_upper and has_digit and has_special:
        return True
    else:
        return False
```
